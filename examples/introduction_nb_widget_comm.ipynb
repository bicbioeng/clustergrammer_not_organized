# Downsample Example

```python
net.load_df(df['clean'], meta_col=df['meta_col'], col_cats=['Gender', 'Category'])
net.downsample(axis='col', num_samples=5, random_state=99, ds_name='DS-Clusters', ds_cluster_name='default')
net.set_manual_category(col='Gender')
net.widget()
```
